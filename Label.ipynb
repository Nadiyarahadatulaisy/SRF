```python
import pandas as pd

# coba dengan encoding latin-1 (sering dipakai kalau ada simbol °)
df = pd.read_csv(file_path, delimiter=";", decimal=",", encoding="latin1")

# pastikan kolom numerik terkonversi
num_cols = ["pH", "Temperatur (°C)", "Kadar Air (%)", "Kelembapan (%)"]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")

# fungsi status
def get_status(row):
    if (6 < row['pH'] < 7) and (20 < row['Temperatur (°C)'] < 30) and (20 < row['Kadar Air (%)'] < 40) and (60 < row['Kelembapan (%)'] < 80):
        return "ok"
    else:
        return "warning"

df["status"] = df.apply(get_status, axis=1)

df.head()

```

```python
output_path = "barubanget_with_status.csv"
df.to_csv(output_path, index=False, encoding="utf-8")

from google.colab import files
files.download(output_path)
```
